```python
bvariable = pd.read_csv('./ResultadoExp/exp2_nuevo.csv')

count = 0
frameList = []


while count < 320:
    datafAux = bvariable[count*10:count*10+10]
    datafAux = datafAux.sort_values(['tiempo'], ascending=[False])
    datafAux = datafAux.drop(datafAux.index[[0,1]])
    datafAux = datafAux.drop(datafAux.index[len(datafAux)-1])
    datafAux = datafAux.drop(datafAux.index[len(datafAux)-1])
    frameList.insert(0, datafAux)
    count+=1
       
bvariable = pd.concat(frameList).reset_index()
bvariable = bvariable.groupby(['Metodo','nroTermIndpte']).mean().reset_index()

count = 0
frameList = []
while count < 32:
    datafAux = bvariable[count*10:count*10+10]
    datafAux = datafAux.sort_values(['tiempo'], ascending=[False])
    datafAux = datafAux.drop(datafAux.index[[0,1]])
    datafAux = datafAux.drop(datafAux.index[len(datafAux)-1])
    datafAux = datafAux.drop(datafAux.index[len(datafAux)-1])
    frameList.insert(0, datafAux)
    count+=1

bvariable = pd.concat(frameList).reset_index()
bvariable = bvariable.groupby(['Metodo','nroTermIndpte']).mean().reset_index()


bvariableC = bvariable[:96]
bvariableG = bvariable[96:]
bvariableG = bvariableG.groupby('sizes').mean().reset_index()
bvariableC = bvariableC.groupby('sizes').mean().reset_index()

xC = bvariableC['sizes']
yC = bvariableC['tiempo']
xG = bvariableG['sizes']
yG = bvariableG['tiempo']
y2 = pow(xG,2)
y3 = 5*pow(xG,3)

fig = plt.figure()
fig.suptitle('Cholesky vs Elim. Gaussiana para distintas b', fontsize=14, fontweight='bold')
ax = fig.add_subplot(111)
fig.subplots_adjust(top=0.85)
fig.patch.set_facecolor('white')
ax.set_xlabel('Tamaño de matriz')
ax.set_ylabel('Tiempo')
plt.plot(xC,yC,'r^' ,label = 'Tiempo CH')
plt.plot(xG,yG, 'bo', label = 'Tiempo EG')
plt.plot(xG,y2, 'g', label = 'O(n^2)')
plt.plot(xG,y3, 'g--', label = '2*O(n^3)')
ax.set_yscale('log')


leg = ax.legend()
leg = plt.legend( loc = 'upper left')
plt.show()



```
