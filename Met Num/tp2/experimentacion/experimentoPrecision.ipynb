```python
# -*- coding: utf-8 -*-

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from pylab import *

plt.style.use('ggplot')
```

CONVERSION DATOS A TABLAS

```python
im3_exp2 = pd.read_csv('../experimentacion/outputs/3im_exp2.csv')
im3_exp2 = im3_exp2.sort_values(['MetodoEleccion','CompPpal','NroPersona'] , ascending=[True, True, True])

im7_exp2 = pd.read_csv('../experimentacion/outputs/7im_exp2.csv')
im7_exp2 = im7_exp2.sort_values(['MetodoEleccion','CompPpal','NroPersona'] , ascending=[True, True, True])
```

Calculo de datos Precision, Hit Rate Y Recall "Mal entrenamiento"

```python
#Vp/((fp+fn)/2)


df_KNN_3 = im3_exp2.loc[im3_exp2.MetodoEleccion =='KNN']
df_V_3 = im3_exp2.loc[im3_exp2.MetodoEleccion =='V']

df_V_3['Precision'] = df_V_3['VerdaderosP']/(df_V_3['VerdaderosP'] + df_V_3['FalsosP'])
df_V_3['Recall'] = df_V_3['VerdaderosP']/(df_V_3['VerdaderosP'] + df_V_3['FalsosN'])
df_V_3 = df_V_3.groupby('CompPpal').mean()
df_V_3 = df_V_3.reset_index()


df_KNN_3['Precision'] = df_KNN_3['VerdaderosP']/(df_KNN_3['VerdaderosP'] + df_KNN_3['FalsosP'])
df_KNN_3['Recall'] = df_KNN_3['VerdaderosP']/(df_KNN_3['VerdaderosP'] + df_KNN_3['FalsosN'])
df_KNN_3 = df_KNN_3.groupby('CompPpal').mean()
df_KNN_3 = df_KNN_3.reset_index()
```

Calculo de datos Precision, Hit Rate y Recall "Buen entrenamiento"

```python
#metodosEleccion = im7_exp2['MetodoEleccion'].unique().tolist()
df_KNN_7 = im7_exp2.loc[im7_exp2.MetodoEleccion =='KNN']
df_V_7 = im7_exp2.loc[im7_exp2.MetodoEleccion =='V']

df_V_7['Precision'] = df_V_7['VerdaderosP']/(df_V_7['VerdaderosP'] + df_V_7['FalsosP'])
df_V_7['Recall'] = df_V_7['VerdaderosP']/(df_V_7['VerdaderosP'] + df_V_7['FalsosN'])
df_V_7 = df_V_7.groupby('CompPpal').mean()
df_V_7 = df_V_7.reset_index()


df_KNN_7['Precision'] = df_KNN_7['VerdaderosP']/(df_KNN_7['VerdaderosP'] + df_KNN_7['FalsosP'])
df_KNN_7['Recall'] = df_KNN_7['VerdaderosP']/(df_KNN_7['VerdaderosP'] + df_KNN_7['FalsosN'])
df_KNN_7 = df_KNN_7.groupby('CompPpal').mean()
df_KNN_7 = df_KNN_7.reset_index()
```

Impresion Data Hit Rate V7 VS KNN7

```python
x_KNN_7_h = df_KNN_7['CompPpal']
y_KNN_7_h = df_KNN_7['VerdaderosP']

x_V_7_h = df_V_7['CompPpal']
y_V_7_h = df_V_7['VerdaderosP']

fig = plt.figure()
fig.suptitle('Medidas de "Hit Rate" con Buen entrenamiento', fontsize=14, fontweight='bold')
ax = fig.add_subplot(111)
fig.subplots_adjust(top=0.85)
fig.patch.set_facecolor('white')
ax.set_xlabel('Componentes Principales')
ax.set_ylabel('Hit Rate')
plt.plot(x_V_7_h,y_V_7_h,'ro' ,label = 'V')
plt.plot(x_KNN_7_h,y_KNN_7_h,'g^' ,label = 'KNN')
leg = ax.legend()
leg = plt.legend( loc = 'lower right')
plt.show()
```

EXPERIMENTO 1

```python
df_exp1 = pd.read_csv('../experimentacion/outputs/exp1.csv')

df_exp1_r = df_exp1.loc[df_exp1.REDNORM =='RED']
df_exp1_n = df_exp1.loc[df_exp1.REDNORM =='NORM']

df_exp1_r_xxt = df_exp1_r.loc[df_exp1.XXTXTX == 0]
df_exp1_r_xtx = df_exp1_r.loc[df_exp1.XXTXTX == 1]

df_exp1_n_xxt = df_exp1_n.loc[df_exp1.XXTXTX == 0]
df_exp1_n_xtx = df_exp1_n.loc[df_exp1.XXTXTX == 1]

count = 0
frameList_r = []
frameList_n = []
while count < 1:
    datafaux_r_xxt = df_exp1_r_xxt[count*10:count*10+20]
    datafaux_r_xtx = df_exp1_r_xtx[count*10:count*10+20]
    datafaux_n_xxt = df_exp1_n_xxt[count*10:count*10+20]
    datafaux_n_xtx = df_exp1_n_xtx[count*10:count*10+20]
    datafaux_r_xxt = datafaux_r_xxt.groupby('REDNORM').mean()
    datafaux_r_xtx = datafaux_r_xtx.groupby('REDNORM').mean()
    datafaux_n_xxt = datafaux_n_xxt.groupby('REDNORM').mean()
    datafaux_n_xtx= datafaux_n_xtx.groupby('REDNORM').mean()
    frameList_r.insert(0, datafaux_r_xxt)
    frameList_n.insert(0, datafaux_n_xxt)
    frameList_r.insert(0, datafaux_r_xtx)
    frameList_n.insert(0, datafaux_n_xtx)

    count+=1

df_exp1_r = pd.concat(frameList_r).reset_index()
df_exp1_n = pd.concat(frameList_n).reset_index()

df_exp1_r.loc[df_exp1_r['XXTXTX'] == 0, 'XXTXTX'] = 'XXT'
df_exp1_r.loc[df_exp1_r['XXTXTX'] == 1, 'XXTXTX'] = 'XTX'
df_exp1_n.loc[df_exp1_n['XXTXTX'] == 0, 'XXTXTX'] = 'XXT'
df_exp1_n.loc[df_exp1_n['XXTXTX'] == 1, 'XXTXTX'] = 'XTX'

print()

       
ax = sns.barplot(x = 'XXTXTX', y = 'tiempo', ci = 0.68, data = df_exp1_r)
ax.set(xlabel='Forma del producto de matrices', ylabel='Tiempo Promedio')
ax.set_title('Comparacion de tiempos para imgenes de tamao reducido')
plt.show()

ax2 = sns.barplot(x = 'XXTXTX', y = 'tiempo', ci = 0.68, data = df_exp1_n)
ax2.set(xlabel='Forma del producto de matrices', ylabel='Tiempo Promedio')
ax2.set_title('Comparacion de tiempos para imgenes de tamao normal')
plt.show()
```
